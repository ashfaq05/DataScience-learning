## 3. What is Machine Learning? 

## 3.1 Basic intuition

Again, like data science, the definition of Machine learning(ML) is up for debate. One of the widely accepted definitions has been provided by Arthur Samuels. 

> Machine learning is the field of study that gives computers the ability to learn without being explicitly programmed. - Arthur Samuels (1959)

First let us understand the difference between a traditional computer program and a machine learning program. We will understand the differences through an example. Suppose we are trying to program how to recognise a car from a set of images. A typical traditional program would help you identify a car through a set of rules. General rules for a car are 
- it has wheels
- it has headlight.
- it has steering wheel

![car_rules](../images/cars_rules.png)

Then for a computer program, you can make the algorithm, 
if it has wheels AND if it has headlight AND if it has steering wheel
Think of all the rules you need to list down to make the computer program work. It would be really CUMBERSOME. Also the program might fail if the images are of a different orientation. 

![car_diff_angles](../images/car_diff.png)

On the other hand, think of a scenario where you can just pass on the different images of car to a program, along with some data to help identify the car and then allow the program to figure out the rest of the rules by itself. This is a typical ML program - which given enough data will generalize and figure out the rest of the rules by itself. 

So the difference between a traditional computer program and ML program is as follows:-
> **Traditional Algorithm** - clear set of instructions to which we give input and get a desired output. A rule-based approach.   
>  **ML Algorithm** - give a set of data and let the algorithm figure out patterns and rules to apply for future inputs.  

Before the advent of ML as a formal field of application and study, rule-based approaches were used to simulate intelligence. The `intelligence` of the computer system which played chess and defeated a grandmaster - [Deep Blue](https://www.scientificamerican.com/article/20-years-after-deep-blue-how-ai-has-advanced-since-conquering-chess/) was essentially improved by a set of rules defined by other grandmasters of chess and improved search algorithms. So, essentially an expert defined a set of rules which helped automate some tasks and reduce human intervention. Thus, another name for these rule-based systems is `expert systems`. 

`Expert systems` worked well because the data generated was very structured in nature and the scope was limited.  Thus, it was plausible to define rules by hand. As the size of data keeps increasing, the rules need to be modified constantly. With the advent of semi-structured and unstructured data, rule-based systems were no longer practical. It would be something like - fixing bugs as the code keeps getting bigger. The advantage of ML is that it gets better as the size of data keeps increasing. In summary, the advantage of ML over expert systems is that it is scalable and generalizes well on data. 

An important thing to understand is that ML provides insights on data and can be thought of as predictive analytics. What it does better and what it can't do w.r.t providing answers is best summarized in this diagram. 

![analytic](../images/analytic-maturity.jpg)

## 3.2 Misconceptions about Machine Learning

Before we deep dive into the various aspects of Machine Learning, it is important to do a sanity check on the popular misconceptions surrounding ML. 

### Myth 1 - Machines can learn autonomously
For the machines to be successful and provide right insights on data, a lot of work needs to be done by human on data to get it ready. True autonomous learning in not possible and human cannot be completely removed from the equation. For example, look at report in Financial Times on how [self-driving cars are proving to be labor intensive](https://www.ft.com/content/36933cfc-620c-11e7-91a7-502f7ee26895). 

### Myth 2 - Machine learning can be applied to any task

> Surprisingly, despite AI’s breadth of impact, the types of it being deployed are still extremely limited. Almost all of AI’s recent progress is through one type, in which some input data (A) is used to quickly generate some simple response (B). - Andrew Ng 2016 (Source: HBR)

Nothing summarizes the response to the myth better than these words of Andrew Ng, one of the pioneers in ML. Read more on his thoughts on this subject [here](https://hbr.org/2016/11/what-artificial-intelligence-can-and-cant-do-right-now). Also, ML cannot be applied effectively when collection of data is very difficult like in health domains. 

On the other side of the coin, a lot of problems in industry can be solved through simple data summarization and analysis and don't need AI. Why use complicated statistical models to solve problems which could be solved by simple rules. 

### Myth 3 - Once data is fed, Machine learning will give transformational insights

A popular myth is that once you feed the algorithm data, actionable insights will emerge like the pot of gold at the edge of the rainbow. It is exactly that - a mirage. For every data generated, defining the problem to solve is very hard and it is a continuous iterative process of refining data and rebuilding models. 

### Myth 4 - Machine Learning will give rise to superhuman intelligence

Machines are as intelligent as the humans who programmed it. Inspite of all the advancement, machines don't have common sense and it cannot be taught the same. Don't worry - your smartphone won't take over the world and enslave you. EVER!

## 4. Types of Machine Learning

## 4.1 Supervised ML

- For ML algorithms, like any other algorithms, you give an input to get an output. For supervised ML, input is tagged with expected outputs and we try to predict the outputs for future inputs too. Some of the examples of supervised learning are as follows:

![supML](../images/ai_supML.png)
Source: [Andrew Ng, HBR](https://hbr.org/2016/11/what-artificial-intelligence-can-and-cant-do-right-now)

- Formally, the main task of the ML algorithm is to learn a mapping function from input to output variable/variables.
$$\mathbf{Y = F(x)}$$

### Train/Test Data

![multiply](../images/child_multiply.jpeg)
Let's take a simple scenario of teaching a child multiplication tables. You have taught everything from $1*1$ and $9*9$. Now how would you test whether the child has truly understood multiplication or simply memorised the tables? A good way to do that would be to give unseen data to test the child like $11*12$. Here we can say that $1*1$ to $9*9$ is the training data and $11*12$ is the test data.

Similarly, in ML too, we would like to see if the ML algorithm has truly learnt the relationship between input and output. To do that, we divide the data into two parts - Training data and Test data. We use training data to learn the given model. And then, we use the test data to evaluate the model. What `learning a model` actually means, we will study it in other concepts down the line. Broadly there are two types of supervised learning tasks - Classification and Regression. 

The main difference between classification and regression is the form of output variable. 
If the output variable is discrete, it is classification, and if it is continuous, it is regression. You can think of discrete data as something which is counted and continuous data as something which is measured. Discrete data refers to things like - movie genres, categories of mail (spam/not spam) and so on. Continuous data refers to data like - height of a person, stock price, price of a house and so on. 

### Classification
- It involves finding a set of categories a new data point belongs to. The output value is a class or a category. 
- Examples
     
     - Is the image - a cat or dog 
     - Deciding if an email is spam or not. 
     - Predicting whether a patient has cancer or not
     - Predicting whether the value of a stock will rise or drop

**Important: **
 - Qualitative Output
 - **Predefined Classes**
         

### Regression
- It involves predicting a continuous value, given a set of inputs. The output value is continuous. 
- Example: 
     - Predict the stock price given different factors 
     - Predicting sales of a company for a given year
     - Predicting runs/goals scored by a sports team
     - Assigning credit rating
     
**Important: **
- Quantitative Output 
- **Previous input-output observations**

![classReg](../images/class_reg.jpg)
Consider a situation, where we have the data of houses in the city.
- Predicting the price of the house, given various factors like area of house, location, no of bedrooms, etc : Regression
- Predicting if house will get sold in the next few months, given various factors like area of house, location, etc: Classification

## 4.2 Unsupervised ML

In the previous chapter, you have learnt about supervised machine learning, where the inputs are tagged with specific outputs. What if, the inputs aren't tagged? Unsupervised machine learning is the machine learning task of inferring a function to describe hidden structure from "unlabeled" data (a classification or categorization is not included in the observations)

 - Unsupervised learning is where you only have input data (X) and no corresponding output variables
 
 - The goal for unsupervised learning is to model the underlying structure or distribution in the data in order to learn more about the data
 
So in this case, we try to find homogeneous groups between data which we refer to as clusters. Typically, objects in a cluster exhibit similar behaviour. 
 
### Clustering 
***
 - Unsupervised Learning Technique
 - Grouping objects in clusters 
    - Similar Within Clusters 
    - Dissimilar between clusters 

![cluster](../images/sup-vs-unsup.png)

Examples: 
 - Segmenting customers into various groups
 - Online shopping recommender system: Grouping things that are bought together
 - Grouping Similar Animal Photos

## 4.3 Semi-supervised ML

- A mix of Supervised and Unsupervised Learning - only little data is labelled and the rest is unlabelled. The difference between the three types of learning: Supervised, Unsupervised and Semi-supervised is seen as below.

![partial_labelling](../images/pseudo-labeling.png)

### Reinforcement Learning

 - Used to solve interacting problems where the data observed up to time "**t**" is considered to *decide which action to take at time* "**t+1**"
 
 - Also used for A.I. when training machines to perform tasks such as Walking, etc. 
 
Agent learns from the environment by interacting with it and receiving reward/punishment. Imagine a child walking towards a fireplace. It will feel warm and go closer towards the fire. This is like reward. When the child goes too close, it will feel hot and go back. This is punishment. Based on reward and punishment, the agent learn to navigate. 

![mario](../images/mario.png)

An agent playing the Mario game is an example of reinforcement learning.

## 5. Mapping the Data Science journey

## 5.1 To ML or not to ML

![ml_check](../images/check_ML.png)

Before we even begin to map the journey of data science, we have to ask the pertinent question - do we need to apply ML to solve our problem? The flowchart shown is a handy guide to decide if the use case needs Machine Learning or not.

Most of the time, we don't realize, but defining the right problem is extremely difficult. The ability to judge whether data science can be applied to a given problem is also an important quality of a data scientist. 

Some important things to ponder on before you apply ML (Source: [Link](https://hbr.org/2016/11/how-to-tell-if-machine-learning-can-solve-your-business-problem)):
- Rule based approach gives you a good accuracy, ML gives you a marginal improvement. 
    - Here, a decision needs to be made if it is worth investing in a ML model. Most of the times, the business costs dictate whether ML is applied or not. 
- Do you require prediction in your problem or causal inference?
    - If the business problem requires investigation of which causes what, then ML is not a good fit. ML is good only at prediction given a host of factors and can't do any causal inference. 
- Is the problem self-contained and insulated by outside influences
    - The data you feed to your learning algorithm includes more or less all there is to the problem. If, in the future, the thing you’re trying to predict changes unexpectedly, and no longer matches prior patterns in the data, the algorithm will not know what to make of it.
  
Good machine learning problems 
   - Predicting the likelihood that a certain type of user will click on a certain kind of ad 
   - Evaluating the extent to which a piece of text is similar to previous texts you have seen.

Problems not fit for ML
   - Predicting profits from the introduction of a completely new and revolutionary product line
   - Extrapolating next year’s sales from past data, when an important new competitor just entered the market.
   
Found a problem where ML could be applied? Congratulations! Now we shall look at the complete workflow of solving a problem. 

## 5.2 ML Workflow

![ml_workflow](../images/ml_workflow2.png)

### Step 1 - Defining the problem 

An often underestimated step in the entire process, but a very crucial step to succeed. It is important to define the scope of the problem well. From the host of possible problems to solve, arriving at the right problem to solve is a valued skill.  


### Step 2 - Prepare Data 

From the vast population of available data, find what are the missing values, and decide which columns are relevant for your processing and remove the rest. Find and engineer new features that you think might help in solving your problem better. 

### Step 3 - Develop the Model

This is the part which every scientist fantasizes to actually do. Develop the model on data and get the insights to answer the business questions defined earlier. 

### Step 4 - Evaluate the Model

Once the model is built, it is important to assess the model and see if it is working well on some test data before it is deployed on unseen data. How to assess the model and choose the best model is also important.
